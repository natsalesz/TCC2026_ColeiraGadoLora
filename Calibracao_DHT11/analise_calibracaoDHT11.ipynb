```python
#agora saber qual(is) sensor(es) realmente apresentam diferenças significativas
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests

# Comparar pares de sensores (com correção para múltiplos testes)
pairs = [("temperatura_sensor_1", "temperatura_sensor_padrao"), ("temperatura_sensor_2", "temperatura_sensor_padrao"), ("temperatura_sensor_1", "temperatura_sensor_2")]
p_values = []

for group1, group2 in pairs:
    stat, p = mannwhitneyu(df[group1], df[group2])
    p_values.append(p)

# Ajustar p-values para evitar falsos positivos (Bonferroni)
reject, adjusted_p = multipletests(p_values, method='bonferroni')[:2]

for (group1, group2), p_adj in zip(pairs, adjusted_p):
    print(f"{group1} vs {group2}: p-ajustado = {p_adj:.4f}")
```

```python
#para o sensor que apresnetou diferença significativa verificar se os residuos apresntam padrao adequado para a determinação da regressão adequada
from statsmodels.stats.stattools import durbin_watson
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = df[['temperatura_sensor_2']]  # Variável independente
y = df['temperatura_sensor_padrao']  # Variável dependente

model = LinearRegression().fit(X, y)
y_pred = model.predict(X)
residuos = y - y_pred

dw = durbin_watson(residuos)
print(f"Durbin-Watson: {dw:.2f}")
```

```python
#apenas para o sensor que apresentou diferença significativa realizar a regressão por meio do Modelo de Regressão Linear Ordinária (OLS)
'''from sklearn.linear_model import LinearRegression

X = df[['temperatura_sensor_2']]  # Variável independente
y = df['temperatura_sensor_padrao']      # Variável alvo (padrão)

model = LinearRegression().fit(X, y)
print(f"Equação de calibração: padrão = {model.intercept_:.2f} + {model.coef_[0]:.2f} * sensor1")'''
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.diagnostic import het_breuschpagan

'''# 2. Gráfico de dispersão
plt.figure(figsize=(8,4))
plt.scatter(df['temperatura_sensor_2'], df['temperatura_sensor_padrao'], alpha=0.5)
plt.title('Relação entre Sensor Aferido e Sensor Baixo Custo')
plt.xlabel('Temperatura Sensor Baixo Custo (°C)')
plt.ylabel('Temperatura Sensor Aferido (°C)')
plt.grid(True)
plt.show()'''

# 3. Modelo de Regressão Linear Ordinária (OLS)
X = df['temperatura_sensor_2']
y = df['temperatura_sensor_padrao']
X = sm.add_constant(X)  # Adiciona intercepto

model_ols = sm.OLS(y, X).fit()
#print("\nResultados da Regressão OLS:\n", model_ols.summary())

# 4. Teste de heterocedasticidade
_, pval, _, _ = het_breuschpagan(model_ols.resid, model_ols.model.exog)
print(f"\nTeste de Breusch-Pagan para heterocedasticidade - p-valor: {pval:.4f}")

# 5. Modelo GLS considerando heterocedasticidade
# Estimando a estrutura de variância dos resíduos
resid_std = np.abs(model_ols.resid)
aux_reg = sm.OLS(np.log(resid_std**2), X).fit()
fitted_values = aux_reg.fittedvalues
weights = 1 / np.exp(fitted_values)

# Ajustando o modelo GLS
model_gls = sm.GLS(y, X, weights=weights).fit()
print("\nResultados da Regressão GLS:\n", model_gls.summary())

# 6. Gráfico dos resíduos do GLS
plt.figure(figsize=(5,3))
plt.scatter(model_gls.fittedvalues, model_gls.resid, alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
#plt.title('GLS Model Residuals vs. Fitted Values')
plt.xlabel('Adjusted Values')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()

# 7. Equação final do modelo
intercept = model_gls.params[0]
slope = model_gls.params[1]
print(f"\nEquação final:  padrao = {intercept:.4f} + {slope:.4f} * sensor2")

# 8. Análise de precisão
mae = np.mean(np.abs(model_gls.resid))
rmse = np.sqrt(np.mean(model_gls.resid**2))
print(f"\nMAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R²: {model_gls.rsquared:.4f}")
```
